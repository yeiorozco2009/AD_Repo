```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from google.colab import drive
```

```python
df = pd.read_excel('/content/gdrive/MyDrive/ANALITICA /0. SolicitantesCrédito.xlsx')
display(df)
```

```python
XD = df.dropna(subset=['PreApr'])
```

```python
#NUEVAS VARIABLES
variables_independientes = ['Edad','Hijos','Perscargo', 'Estrato','Ingresos','Egresos', 'Monto (EAD)', 'Plazo','Cuota (COP)']
```

```python
XD = XD[variables_independientes]
print("\nMatriz de variables independientes (XD):")
display(XD.head())
```

```python
yd = df.dropna(subset=['PreApr'])['PreApr']
```

```python
XDArray = np.array(XD)
ydArray = np.array(yd)
```

```python
modeloNB = GaussianNB()
modeloNB.fit(XDArray, ydArray)
```

```python
ydPred = modeloNB.predict(XDArray)
cm = confusion_matrix(ydArray, ydPred)
```

```python
#METRICAS
VN = cm[0,0]
FP = cm[0,1]
FN = cm[1,0]
VP = cm[1,1]
```

```python

exactitud = (VP +VN)/ cm.sum()
```

```python
error = (FP + FN)/ cm.sum()
```

```python
presicion = VP / (VP + FP)
```

```python
sensibilidad = VP / (VP + FN)
```

```python
especificidad = VN / (VN + FP)
```

```python
print(f'la exactitud del modelo es de {exactitud:.2%}')
print(f'la error del modelo es de {error:.2%}')
print(f'la presicion del modelo es de {presicion:.2%}')
print(f'la sensibilidad del modelo es de {sensibilidad:.2%}')
```

EPLICACIÓN DE LOS NUEVOS RESULTADOS :

Al comparar el modelo con los resultados del anterior, se puede observar que el rendimiento general se vio afectado por la inclusión de las nuevas variables. Al introducir datos como "monto, el plazo y la cuota al mismo tiempo, se viola el principio de naive bayes (ya que las variables fucionan mejor cuando son totalmente independientes)

Por otra parte, las predicciones especificas, observamos algo interesante entre, la presición y la sensibilidad. la precisión mejoro, lo que inidica que el modelo es confiable cuando decide aprobar un credito. No obstante la sensibilidad cayó al 63%, lo que nos indica que el modelo se ha vuelto demasiado estricto. Esto puede pasar ya que las variables demográficas adicionales, como el "numero de hijos" o "personas a cargo" introducen ruido, probocando que el algoritmo prefiera rechazar a un cliente ante la duda, antes que arriesgarse a aprobarlo incorrectamente.
