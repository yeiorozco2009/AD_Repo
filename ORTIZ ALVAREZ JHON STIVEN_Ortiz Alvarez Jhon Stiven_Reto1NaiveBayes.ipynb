```python
## Escojo la variable dependiente
yd = solicitantes["PreApr"].dropna()
yd
```

### Implementamos el modelo Naive Bayes

```python
### Importar las librerias necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.naive_bayes import GaussianNB
```

```python
### Se simplifica ahora, volviendolo un array (se necesita solo texto, no tan interactivo)
XDArray = np.array(XD)
ydArray = np.array(yd)
```

```python
### Entrenamos el modelo (Fase Train)
modeloNB = GaussianNB() ## Variable que aloja el modelo
modeloNB.fit(XDArray, ydArray)
```

```python
## Predecimos los yd (Etapa test)
ydPred = modeloNB.predict(XDArray)      #Esto es lo que predice el modelo (diferente de los reales)
# ctrl + shift + enter = impresión selectiva
```

## Ahora se ejecuta la matriz de confusión, paso #4

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(ydArray, ydPred)    #Confusion Matrix
cm
```

```python
### #5 Metricas de evalución del modelo
```

```python
## Extraer valores de la matrix de confusión
VN = cm[0,0]; FP = cm[0,1]; FN = cm[1,0]; VP = cm[1,1]
```

```python
## Exactitud
exactitud = (VN + VP)/cm.sum()

## Error
error = (FP + FN)/cm.sum()

## Precisión
precisión = VP/(FP+VP)

## Sensibilidad
sensibilidad = VP/(VP + FN)

## Especificidad
especifidad = VN/(VN + FP)
```

```python
## Imprimo los mensajes
print(f"La exactitud del modelo es de{exactitud: .2%}")
print(f"El margen de error del modelo es de{error: .2%}")
print(f"La precisión del modelo es de{precisión: .2%}")
print(f"La sensibilidad del modelo es de{sensibilidad: .2%}")
print(f"La especifidad del modelo es de{especifidad: .2%}")
```

## - La exactitud desmejoró, esto puede ser explicado a través de predicciones menos correctas por mayor volumen
##- El margen de error subió, un resultado directo de una disminución en la exactitud
## - La precición subió en poco porcentaje, mostrando un incremento casi que insignificante en la predicción de preaprobados
##- La sensibilidad disminuyó, mostrando un peor desempeño en predecir los realmente aprobados
## - La especificidad aumentó ligeramente, prediciendo un poco mejor los realmente prenegados

### 6. Generalización del modelo

```python
# Las convierto en arreglo
XDGenerArray = np.array(XDGener)
```
