IMPLEMENTAMOS EL MODELO NAIVE BAYES


```python
### IMPLEMENTAMOS LAS LIBRERIAS NECESARIAS
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import GaussianNB
import numpy as np
import pandas as pd

def metricas_binarias_desde_cm(cm):
    """Calcula métricas a partir de una matriz de confusión 2x2:
    [[VN, FP],
     [FN, VP]]
    """
    VN, FP, FN, VP = cm[0,0], cm[0,1], cm[1,0], cm[1,1]
    total = VN + FP + FN + VP

    exactitud = (VP + VN) / total if total else np.nan
    error = (FP + FN) / total if total else np.nan
    sensibilidad = VP / (VP + FN) if (VP + FN) else np.nan   # recall
    precision = VP / (VP + FP) if (VP + FP) else np.nan
    especificidad = VN / (VN + FP) if (VN + FP) else np.nan

    return {
        'VN': int(VN), 'FP': int(FP), 'FN': int(FN), 'VP': int(VP),
        'Exactitud': exactitud,
        'Error': error,
        'Precision': precision,
        'Sensibilidad': sensibilidad,
        'Especificidad': especificidad,
    }

def entrenar_y_evaluar(nombre, X, y):
    X_df = pd.DataFrame(X).copy()
    y_sr = pd.Series(y, name="y").copy()

    data = pd.concat([X_df, y_sr], axis=1)

    data = data.dropna()

    X_clean = data.iloc[:, :-1].to_numpy()
    y_clean = data.iloc[:, -1].to_numpy()

    y_clean = y_clean.astype(int)

    modelo = GaussianNB()
    modelo.fit(X_clean, y_clean)

    y_pred = modelo.predict(X_clean)  # mismo conjunto
    cm = confusion_matrix(y_clean, y_pred)

    metr = metricas_binarias_desde_cm(cm)
    metr['Modelo'] = nombre
    metr['Filas'] = len(y_clean)
    return modelo, y_pred, cm, metr

modelo_old, ypred_old, cm_old, metr_old = entrenar_y_evaluar('Anterior (4 vars)', XD_old, yd_old)
modelo_new, ypred_new, cm_new, metr_new = entrenar_y_evaluar('Reto 1 (9 vars)', XD_new, yd_new)


modeloNB = modelo_new
ydPred = ypred_new
ydArray = np.array(yd_new)
cm = cm_new


comparacion = pd.DataFrame([metr_old, metr_new]).set_index('Modelo')[
    ['Filas','VN','FP','FN','VP','Exactitud','Error','Precision','Sensibilidad','Especificidad']
]
comparacion
```

En este ejercicio se re-entrenó un modelo de Gaussian Naive Bayes para la
predicción de la variable PreApr, ampliando el conjunto de variables
independientes de 4 a 9, incluyendo variables demográficas y financieras:
Edad, Hijos, Personas a cargo, Estrato, Ingresos, Egresos, Monto (EAD),
Plazo y Cuota (COP).


Los resultados obtenidos muestran que el modelo con un mayor número de
variables no presenta una mejora en el desempeño global frente al modelo
anterior. En particular, la exactitud disminuyó de 0.7614 a 0.7499 y la
sensibilidad se redujo de 0.6746 a 0.6389, indicando una menor capacidad
del modelo para identificar correctamente las solicitudes aprobadas.
Aunque se observa una leve mejora en la precisión y la especificidad,
estos incrementos no compensan la pérdida en capacidad de predicción.

Esto sugiere que agregar nuevas variables
financieras introduce correlaciones que afectan el supuesto de independencia condicional propio del modelo Gaussian Naive
Bayes. Por eso se puede concluir que un modelo más
simple con menos variables resulta más efectivo para este caso.
