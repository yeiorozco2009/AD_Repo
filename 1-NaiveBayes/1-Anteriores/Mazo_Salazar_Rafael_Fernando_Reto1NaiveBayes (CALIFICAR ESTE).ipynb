```python
### 3. Eliminamos los registros donde PreAp esta vacio
XD = Solicitantes.dropna(subset=['PreApr'])
```

```python
Solicitantes.info()
```

```python
## Escojo solo las variables que voy a usar como independiente.
XD = XD[["Edad", "Hijos", "Perscargo", "Estrato", "Ingresos", "Egresos", "Monto (EAD)", "Plazo", "Cuota (COP)"]]
```

```python
## Escojemos las variables independientes
yd = Solicitantes["PreApr"].dropna()
```

```python
### Importamos las librerias que necesitamos
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.naive_bayes import GaussianNB
```

```python
### Convierto los dataset a array
XDArray = np.array(XD)
ydArray = np.array(yd)
```

```python
### ENTRENAMOS EL MODELO (FASE TRAIN)
modeloNB = GaussianNB()
modeloNB.fit(XDArray, ydArray)
```

```python
## Predecimos los yd (Etapa test)
ydPred = modeloNB.predict(XDArray)
```

```python
### Matriz de confusion
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(ydArray, ydPred)
cm
```

```python
## Extraer los valores de la matriz de confusión
VN = cm[0,0]; FP = cm[0,1]; FN = cm[1,0]; VP = cm[1,1]
```

```python
## Exactitud
exactitud = (VP+VN)/ cm.sum()

## Error
error = (FP+FN)/ cm.sum()

## Precisión
precision = VP/(FP+VP)

## Sensibilidad
sensibilidad = VP/(VP+FN)

## Especificidad
especificidad = VN/(FP+VN)
```

```python
## Imprimo la Información
print(f'La Exactitud del modelo es:     {exactitud:.2%}')
print(f'El Error del modelo es:         {error:.2%}')
print(f'La Precisión del modelo es:     {precision:.2%}')
print(f'La sensibilidad del modelo es:  {sensibilidad:.2%}')
print(f'La Especificidad del modelo es: {especificidad:.2%}')
```

```python
# Filtro los que no tienen clasificacion (No tienen valor en la columna PreApr)
XDGener = Solicitantes[Solicitantes["PreApr"].isna()]
```

```python
## Filtro las variables
XDGener = XDGener[["Edad", "Hijos", "Perscargo", "Estrato", "Ingresos", "Egresos", "Monto (EAD)", "Plazo", "Cuota (COP)"]]
```

```python
# Las convierto en Arreglos
XDGenerArray = np.array(XDGener)
```

```python
# Generalizar (Las predicciones para los solicitantes faltantes)
ydGner = modeloNB.predict(XDGenerArray)
```

```python
ydGenerSerie = pd.Series(ydGner, index= Solicitantes[Solicitantes["PreApr"].isna()].index )
ydGenerSerie
```

```python
### Completamos la base de datos Principal
Solicitantes["PreApr"].fillna(ydGenerSerie, inplace=True)
```

Esto ocurre porque el modelo logra identificar mejor los casos positivos, reduciendo los falsos negativos y mejorando la sensibilidad. Es decir, se le escapan menos casos importantes que antes por tener mas informacion.

Sin embargo, esta mejora hace que el modelo también cometa un poco más de falsos positivos, por lo que la especificidad y la precisión bajan levemente. el modelo es mejor detectando positivos, pero a veces se equivoca marcando como positivo algo que no lo es.

En conclusión, agregar más variables ayuda a que el modelo entienda mejor los datos, pero también cambia el equilibrio entre aciertos y errores. Por eso, no siempre más variables significa un modelo perfecto, sino uno que debe evaluarse según el objetivo del análisis.
