## Implementamos el modelo de NAYBE BAYES

```python
## iMPORTAMOS LAS LIBRERIAS NECESARIAS
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.naive_bayes import GaussianNB
```

```python
### Conviertes las datasets a arrays
XDArray = np.array(XD)
ydArray = np.array(yd)
```

```python
### ENTRENAMOS EL MODELO (Fase Train)
modeloNB = GaussianNB() ## variable que aloja el modelo
modeloNB.fit(XDArray, ydArray)
```

```python
## Predecimos los yd (Etapa test)
ydPred = modeloNB.predict(XDArray)
```

### Matriz de confusión

```python
from sklearn.metrics import confusion_matrix
## creamos una variable para contener la matriz de confusión
```

```python
cm = confusion_matrix(ydArray, ydPred)
cm
```

## Calcular las métricas de la evaluación del módelo

```python
## Extraigo los valores de la matriz de confusión
VN = cm[0,0] ; FP = cm[0,1]; FN = cm[1,0]; VP = cm[1,1]
```

```python
## EXACTITUD
exactitud = (VP + VN) / cm.sum()
print(f"La exactitud del modelo es de: {exactitud}")

## ERROR
error = (FP + FN) / cm.sum()
print(f"El error del modelo es de: {error}")

## SENSIBILIDAD
sensibilidad = VP / (VP + FN)
print(f"La sensibilidad del modelo es de: {sensibilidad}")

## ESPECIFICIDAD
especificidad = VN / (VN + FP)
print(f"La especificidad del modelo es de: {especificidad}")

## PRECISION
precision = VP / (VP + FP)
print(f"La precisión del modelo es de: {precision}")
```

```python
## Imprimo los mensajes
```

```python
print(f"La exactitud del modelo es de: {exactitud:.2%}")
#El modelo presenta un desempeño aceptable para un problema real de riesgo crediticio,
#considerando la complejidad del comportamiento financiero.
print(f"El error del modelo es de: {error:.2%}")
#Aunque existe margen de mejora, el error se mantiene en un nivel razonable para un primer modelo de clasificación.
print(f"La sensibilidad del modelo es de: {sensibilidad:.2%}")
#De todos los solicitantes que realmente eran preaprobados, el modelo identifica correctamente cerca del 64 %.
print(f"La especificidad del modelo es de: {especificidad:.2%}")
#De todos los solicitantes que realmente debían ser rechazados, el modelo identifica correctamente casi el 86 %
print(f"La precisión del modelo es de: {precision:.2%}")
#Cuando el modelo dice que un solicitante es preaprobado, acierta en más del 81 % de los casos.
```
